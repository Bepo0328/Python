### 여기서 출제자의 의도가 조건문과 함수 구현이라면??
- Q. 모 기업의 입사 시험은 심사를 보고, 면접자의 점수를 매긴다. 심사는 공정성을 위해서 심사 점수의 <span style="color:green">최대값과 최솟값을 제외한 합산</span>을 구한다. 단, 심사위원이 <span style="color:yellow">2명 이하인</span> 경우는 제외하지 않고, 점수를 합산한다.


- 개인의 점수는 <span style="color:green">소수 나머지를 버린 합산의 평균</span>으로 주어지며, <span style="color:red">상위 K명</span>만 면접을 통과한다. 상위 <span style="color:yellow">K번째 점수와 동점인 사람은 모두 합격</span> 처리한다.(평균은 제외한 2명을 세지 않는다.)


- 면접자 수빈이는 실수로 채점표를 볼 수 있었고, 본인의 점수를 포함한 모든 참가자의 점수를 확인하고 수정할 수 있었다. 수빈이는 본인의 심사표를 수정하여, 면접을 통과하고자 한다. 이 경우, 심사표의 점수를 <span style="color:red">최소 몇 점 수정해야 통과</span>할 수 있을까? 수정이란 <span style="color:red">기존 점수와 바꾼 점수의 차</span>를 의미한다. 또한 아직 최대와 최소를 제외하지 않았기에 <span style="color:yellow">수정한 점수에서 최대와 최소를 빼게</span> 된다.


- 참가자는 <span style="color:red">1번부터 N번$(<100,000)$</span>까지이고, 수빈이는 참가 번호 1번이다. 참가자에 따라 <span style="color:red">심사위원 수$(<=10)$</span>는 다를 수 있다. 심사위원 당 점수를 줄 수 있는 최대점은 100점이다.

- <span style="color:gray">N이하의 소수 리스트를 반환하는 함수</span>
```python
def era(N):
    ck = [False for i in range(N + 1)]
    ret = []
    for i in range(2, N + 1):
        if ck[i]: continue
        ret.append(i)
        for j in range(i ** 2, N + 1, i): ck[j] = True
     #print(ret)
    return ret
```
